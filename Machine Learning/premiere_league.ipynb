```python
import pandas as pd

df = pd.read_excel('premiere_transposta.xlsx')
df.head()
```

```python
# Criando uma coluna 'resultado' usando Lambda, nÃ£o criei pelo excel pois aqui atualiza automaticamente quando atualizar a planilha.

df['resultado'] = df.apply(
  lambda row: 'vitoria_casa' if row['gols_casa'] > row['gols_fora'] else 'vitoria_visitante' if row['gols_fora'] > row['gols_casa'] else 'empate', axis=1 )

df.head()
```

```python
# Codificando as colunas chaves usando o one-hot encoding, usando o metodo dummies do pandas

df = pd.get_dummies(df, columns=['clube', 'time_casa', 'time_fora'])

df.head()
```

```python
# preenchendo os dados NaN com uma mediana
#dados = df.fillna(df.median(numeric_only=True))
#dados
```

```python
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import LabelEncoder
import xgboost as xgb

# dividindo em treino e teste
# Target Ã© o resultado, e para saber o resultado, precisamos saber o numero de gols de cada time.
# Foi dropado tbm as colunas gols, para nÃ£o dar vazamento de dados

x = df.drop(columns=['resultado'])
y = df['resultado']

# criando treino e teste
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=.2, random_state=42)


#label_encoder = LabelEncoder()
#Aqui, um objeto da classe LabelEncoder Ã© criado. O LabelEncoder Ã© uma classe da biblioteca sklearn.preprocessing usada para converter variÃ¡veis categÃ³ricas (ou seja, com valores nÃ£o numÃ©ricos) em valores numÃ©ricos. Cada categoria Ãºnica serÃ¡ mapeada para um nÃºmero distinto.

#Loop sobre colunas de tipo 'object' em X_train:

#for col in X_train.select_dtypes(include=['object']).columns:
#Neste passo, o cÃ³digo estÃ¡ iterando sobre todas as colunas de X_train que sÃ£o do tipo object, ou seja, aquelas que contÃªm dados categÃ³ricos (strings ou outras representaÃ§Ãµes de categorias). A funÃ§Ã£o select_dtypes Ã© usada para filtrar apenas as colunas do tipo 'object'.

#Concatenando os dados de treino e teste:

#python
#cÃ³digo
#combined = pd.concat([X_train[col], X_test[col]], axis=0)
#Para garantir que o LabelEncoder tenha visibilidade de todas as categorias possÃ­veis, tanto no conjunto de treino quanto no de teste, as colunas col de X_train e X_test sÃ£o concatenadas. Isso cria uma Ãºnica sÃ©rie (combined) contendo as categorias de ambas as fontes de dados. O axis=0 indica que a concatenaÃ§Ã£o deve ocorrer verticalmente, ou seja, empilhando as linhas.

#Treinando o LabelEncoder:

##Copiar cÃ³digo
#label_encoder.fit(combined)
#O fit() do LabelEncoder Ã© usado para aprender os mapeamentos entre as categorias e os nÃºmeros. A funÃ§Ã£o recebe os dados concatenados combined e identifica todas as categorias Ãºnicas. Cada categoria serÃ¡ atribuÃ­da a um nÃºmero inteiro.

#Transformando os dados de treino e teste:

#python
#Copiar cÃ³digo
#X_train[col] = label_encoder.transform(X_train[col])
#X_test[col] = label_encoder.transform(X_test[col])
#ApÃ³s o fit(), o transform() Ã© chamado para converter as categorias em nÃºmeros. O transform() Ã© aplicado tanto nas colunas de treino quanto de teste para que os dados sejam convertidos de forma consistente. Os resultados sÃ£o atribuÃ­dos de volta Ã s colunas originais de X_train e X_test.

#X_train[col] Ã© transformado para substituir as categorias por nÃºmeros inteiros.
#O mesmo Ã© feito para X_test[col], garantindo que o modelo veja as mesmas categorias numericamente tanto no treino quanto no teste.


label_encoder = LabelEncoder()
for col in X_train.select_dtypes(include=['object']).columns:
    # Concatenar treino e teste para garantir que o encoder veja todos os valores
    combined = pd.concat([X_train[col], X_test[col]], axis=0)
    label_encoder.fit(combined)
    
    X_train[col] = label_encoder.transform(X_train[col])
    X_test[col] = label_encoder.transform(X_test[col])
```

```python
#le = LabelEncoder(): Cria uma instÃ¢ncia do LabelEncoder. Esse objeto serÃ¡ responsÃ¡vel por transformar as variÃ¡veis categÃ³ricas em valores numÃ©ricos.
#y_train_encoded = le.fit_transform(y_train): Aplica a transformaÃ§Ã£o na variÃ¡vel alvo (y_train), que normalmente contÃ©m classes ou rÃ³tulos categÃ³ricos (por exemplo, "vitÃ³ria", "empate", "derrota"). O mÃ©todo fit_transform() aprende as classes presentes em y_train e as mapeia para valores inteiros. Por exemplo, se vocÃª tiver "vitÃ³ria", "derrota" e "empate", elas podem ser convertidas para 0, 1 e 2, respectivamente.

# Instanciando o label
le = LabelEncoder()
y_train_encoded = le.fit_transform(y_train)
```

```python
#InstanciaÃ§Ã£o do modelo XGBClassifier:

##Copiar cÃ³digo
#xgb = xgb.XGBClassifier(booster='gbtree')
#Aqui, Ã© criado um objeto do classificador XGBClassifier da biblioteca xgboost. Esse classificador Ã© utilizado para modelos de aprendizado de mÃ¡quina baseados em Ã¡rvores de decisÃ£o, usando o algoritmo Gradient Boosting.
#O parÃ¢metro booster='gbtree' especifica que o modelo usarÃ¡ Ã¡rvores de decisÃ£o como o "booster" (tipo de modelo base). Outros tipos possÃ­veis de booster sÃ£o gblinear (para modelos lineares) e dart (Dropouts meet Multiple Additive Regression Trees).
#Realizando a validaÃ§Ã£o cruzada:

##opiar cÃ³digo
##cross_val_score() Ã© uma funÃ§Ã£o da biblioteca sklearn.model_selection que realiza a validaÃ§Ã£o cruzada. Ela avalia o desempenho do modelo em vÃ¡rias divisÃµes do conjunto de dados (em geral, Ã© feito um "k-fold cross-validation").
#xgb Ã© o modelo que serÃ¡ avaliado.
#X_train sÃ£o os dados de entrada (variÃ¡veis independentes), e y_train_encoded sÃ£o as variÃ¡veis alvo (ou seja, as classes ou rÃ³tulos).
#O cross_val_score() retorna um array de escores, que sÃ£o a mÃ©trica de avaliaÃ§Ã£o (geralmente a acurÃ¡cia) para cada divisÃ£o do conjunto de dados de validaÃ§Ã£o cruzada.
#O resultado Ã© multiplicado por 100 para transformar os escores em porcentagens, jÃ¡ que o cross_val_score() normalmente retorna valores entre 0 e 1.
#Calculando a pontuaÃ§Ã£o mÃ©dia:

#python
#Copiar cÃ³digo
#average_score = scores.mean()
#O mÃ©todo .mean() calcula a mÃ©dia dos escores obtidos durante a validaÃ§Ã£o cruzada.
#O valor resultante serÃ¡ a mÃ©dia da acurÃ¡cia ou outra mÃ©trica de avaliaÃ§Ã£o que foi usada pelo cross_val_score.
#Exibindo o resultado:

#ython
#Copiar cÃ³digo
#print(f"PontuaÃ§Ã£o mÃ©dia de validaÃ§Ã£o cruzada: {average_score:.2f}%")
#O resultado final Ã© impresso no formato de uma string formatada.
#{average_score:.2f} significa que o valor da mÃ©dia serÃ¡ exibido com duas casas decimais.
#O print() irÃ¡ exibir a pontuaÃ§Ã£o mÃ©dia da validaÃ§Ã£o cruzada em porcentagem (pois a multiplicaÃ§Ã£o por 100 foi feita anteriormente).

xgb = xgb.XGBClassifier(booster='gbtree')
scores = 100 * cross_val_score(xgb, X_train, y_train_encoded)
average_score = scores.mean()
print(f"PontuaÃ§Ã£o mÃ©dia de validaÃ§Ã£o cruzada:{average_score:.2f}%")
```

```python
#Treinamento do modelo:

##Copiar cÃ³digo
#classificador_xgb_gblinear.fit(X_train, y_train_encoded)
#Aqui, o modelo classificador_xgb_gblinear estÃ¡ sendo treinado usando os dados de treinamento X_train e os rÃ³tulos codificados y_train_encoded.
#O mÃ©todo .fit() Ã© utilizado para treinar o modelo. Ele recebe dois argumentos:
#X_train: As variÃ¡veis independentes (tambÃ©m chamadas de caracterÃ­sticas ou atributos), ou seja, os dados de entrada para o treinamento.
#y_train_encoded: Os rÃ³tulos ou classes (variÃ¡veis dependentes) que correspondem aos dados em X_train. Neste caso, eles jÃ¡ estÃ£o codificados de forma numÃ©rica, o que Ã© necessÃ¡rio para o modelo funcionar corretamente.
#O treinamento envolve ajustar os parÃ¢metros do modelo com base nos dados de entrada para aprender a relaÃ§Ã£o entre as variÃ¡veis de entrada e os rÃ³tulos.
#O classificador_xgb_gblinear provavelmente Ã© uma instÃ¢ncia do XGBClassifier com o booster definido como 'gblinear' (em vez de 'gbtree', que Ã© o padrÃ£o). Isso significa que o modelo usarÃ¡ gradiente boosting com modelos lineares em vez de Ã¡rvores de decisÃ£o.

#Fazendo previsÃµes com o modelo treinado:

#python
#Copiar cÃ³digo
#predict = classificador_xgb_gblinear.predict(X_test)
#ApÃ³s o modelo ser treinado, o mÃ©todo .predict() Ã© usado para fazer previsÃµes no conjunto de dados de teste, X_test.
#O modelo utiliza as informaÃ§Ãµes aprendidas durante o treinamento para prever os rÃ³tulos ou classes para os dados de teste. X_test contÃ©m as mesmas caracterÃ­sticas de entrada que X_train, mas sem os rÃ³tulos (que o modelo precisa prever).
#O resultado de .predict() serÃ¡ armazenado na variÃ¡vel predict, que serÃ¡ um array contendo as previsÃµes feitas pelo modelo para cada instÃ¢ncia em X_test.

xgb.fit(X_train, y_train_encoded)
predict = xgb.predict(X_test)
```

```python
# Converter os valores preditos para os nomes das classes
predict_labels = le.inverse_transform(predict)

# Criar o DataFrame com as classes reais e preditas
resultado = pd.DataFrame({
    'Real': y_test.values,  # Valores reais jÃ¡ estÃ£o com os nomes originais
    'Predito': predict_labels  # Converter para os nomes das classes
})
resultado
```

```python
import pandas as pd
import joblib
from sklearn.preprocessing import LabelEncoder

# 1. Carregar o modelo treinado
modelo_carregado = joblib.load('modelo_premiere02.pkl')

# 2. Carregar os novos dados
dados_novos = pd.read_excel('previsao_premiere_para_teste.xlsx')

# 3. Verificar as colunas presentes no DataFrame
print("Colunas disponÃ­veis no arquivo:", dados_novos.columns)

# 4. Copiar colunas relevantes para referÃªncia posterior
times_casa = dados_novos.get('time_casa', None)
times_fora = dados_novos.get('time_fora', None)

# 5. Remover colunas irrelevantes
colunas_irrelevantes = ['Last 5', 'jogador', 'club', 'jogador_expected_goals', 'team', 'data']
dados_novos = dados_novos.drop(columns=[col for col in colunas_irrelevantes if col in dados_novos.columns], errors='ignore')

# 6. Converter colunas categÃ³ricas para numÃ©ricas usando LabelEncoder
for col in dados_novos.select_dtypes(include=['object']).columns:
    dados_novos[col] = LabelEncoder().fit_transform(dados_novos[col])

# 7. Verificar se todas as colunas estÃ£o no formato numÃ©rico
print("\nTipos de dados apÃ³s prÃ©-processamento:")
print(dados_novos.dtypes)

# 8. Garantir que as colunas correspondam ao modelo
colunas_modelo = modelo_carregado.feature_names_in_

# Adicionar colunas faltantes no DataFrame com valor 0
colunas_faltando = [coluna for coluna in colunas_modelo if coluna not in dados_novos.columns]
for coluna in colunas_faltando:
    dados_novos[coluna] = 0

# Remover colunas extras que nÃ£o sÃ£o esperadas pelo modelo
dados_novos = dados_novos[colunas_modelo]

# 9. Realizar as previsÃµes
try:
    y_pred = modelo_carregado.predict(dados_novos)
except Exception as e:
    print("\nErro durante a previsÃ£o:", str(e))
    raise

# 10. Adicionar as previsÃµes ao DataFrame
dados_novos['PrevisÃ£o'] = y_pred

# 11. Reintroduzir as colunas originais e a data (se disponÃ­veis)
if times_casa is not None:
    dados_novos['time_casa'] = times_casa
if times_fora is not None:
    dados_novos['time_fora'] = times_fora

# Reintroduzir a coluna 'data', se presente no arquivo original
if 'data' in pd.read_excel('bundesliga_para_teste.xlsx').columns:
    dados_novos['data'] = pd.read_excel('bundesliga_para_teste.xlsx')['data']
else:
    dados_novos['data'] = None

# 12. Mapeamento das previsÃµes
mapeamento_previsoes = {
    0: "vitÃ³ria_casa",
    1: "empate",
    2: "vitÃ³ria_visitante"
}
dados_novos['PrevisÃ£o'] = dados_novos['PrevisÃ£o'].map(mapeamento_previsoes)

# 13. Exibir as previsÃµes
print("\nPrevisÃµes de resultados para os novos dados:")
for i, (pred, data, time_casa, time_fora) in enumerate(zip(
        dados_novos['PrevisÃ£o'], 
        dados_novos.get('data', []), 
        dados_novos['time_casa'], 
        dados_novos['time_fora']
    )):
    print(f"Linha {i+1}: PrevisÃ£o - {pred}, Data - {data}, Time Casa - {time_casa}, Time Fora - {time_fora}")

# 14. Opcional: Salvar as previsÃµes em um arquivo Excel
dados_novos.to_excel('previsoes_bundesliga.xlsx', index=False)
print("\nAs previsÃµes foram salvas em 'previsoes_bundesliga.xlsx'.")
```
