```python
import pandas as pd

df = pd.read_excel('premiere_transposta.xlsx')
df.head()
```

```python
# Criando uma coluna 'resultado' usando Lambda, não criei pelo excel pois aqui atualiza automaticamente quando atualizar a planilha.

df['resultado'] = df.apply(
  lambda row: 'vitoria_casa' if row['gols_casa'] > row['gols_fora'] else 'vitoria_visitante' if row['gols_fora'] > row['gols_casa'] else 'empate', axis=1 )

df.head()
```

```python
# Codificando as colunas chaves usando o one-hot encoding, usando o metodo dummies do pandas

df = pd.get_dummies(df, columns=['clube', 'time_casa', 'time_fora'])

df.head()
```

```python
# preenchendo os dados NaN com uma mediana
#dados = df.fillna(df.median(numeric_only=True))
#dados
```

```python
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import LabelEncoder
import xgboost as xgb

# dividindo em treino e teste
# Target é o resultado, e para saber o resultado, precisamos saber o numero de gols de cada time.
# Foi dropado tbm as colunas gols, para não dar vazamento de dados

x = df.drop(columns=['resultado'])
y = df['resultado']

# criando treino e teste
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=.2, random_state=42)


#label_encoder = LabelEncoder()
#Aqui, um objeto da classe LabelEncoder é criado. O LabelEncoder é uma classe da biblioteca sklearn.preprocessing usada para converter variáveis categóricas (ou seja, com valores não numéricos) em valores numéricos. Cada categoria única será mapeada para um número distinto.

#Loop sobre colunas de tipo 'object' em X_train:

#for col in X_train.select_dtypes(include=['object']).columns:
#Neste passo, o código está iterando sobre todas as colunas de X_train que são do tipo object, ou seja, aquelas que contêm dados categóricos (strings ou outras representações de categorias). A função select_dtypes é usada para filtrar apenas as colunas do tipo 'object'.

#Concatenando os dados de treino e teste:

#python
#código
#combined = pd.concat([X_train[col], X_test[col]], axis=0)
#Para garantir que o LabelEncoder tenha visibilidade de todas as categorias possíveis, tanto no conjunto de treino quanto no de teste, as colunas col de X_train e X_test são concatenadas. Isso cria uma única série (combined) contendo as categorias de ambas as fontes de dados. O axis=0 indica que a concatenação deve ocorrer verticalmente, ou seja, empilhando as linhas.

#Treinando o LabelEncoder:

##Copiar código
#label_encoder.fit(combined)
#O fit() do LabelEncoder é usado para aprender os mapeamentos entre as categorias e os números. A função recebe os dados concatenados combined e identifica todas as categorias únicas. Cada categoria será atribuída a um número inteiro.

#Transformando os dados de treino e teste:

#python
#Copiar código
#X_train[col] = label_encoder.transform(X_train[col])
#X_test[col] = label_encoder.transform(X_test[col])
#Após o fit(), o transform() é chamado para converter as categorias em números. O transform() é aplicado tanto nas colunas de treino quanto de teste para que os dados sejam convertidos de forma consistente. Os resultados são atribuídos de volta às colunas originais de X_train e X_test.

#X_train[col] é transformado para substituir as categorias por números inteiros.
#O mesmo é feito para X_test[col], garantindo que o modelo veja as mesmas categorias numericamente tanto no treino quanto no teste.


label_encoder = LabelEncoder()
for col in X_train.select_dtypes(include=['object']).columns:
    # Concatenar treino e teste para garantir que o encoder veja todos os valores
    combined = pd.concat([X_train[col], X_test[col]], axis=0)
    label_encoder.fit(combined)
    
    X_train[col] = label_encoder.transform(X_train[col])
    X_test[col] = label_encoder.transform(X_test[col])
```

```python
#le = LabelEncoder(): Cria uma instância do LabelEncoder. Esse objeto será responsável por transformar as variáveis categóricas em valores numéricos.
#y_train_encoded = le.fit_transform(y_train): Aplica a transformação na variável alvo (y_train), que normalmente contém classes ou rótulos categóricos (por exemplo, "vitória", "empate", "derrota"). O método fit_transform() aprende as classes presentes em y_train e as mapeia para valores inteiros. Por exemplo, se você tiver "vitória", "derrota" e "empate", elas podem ser convertidas para 0, 1 e 2, respectivamente.

# Instanciando o label
le = LabelEncoder()
y_train_encoded = le.fit_transform(y_train)
```

```python
#Instanciação do modelo XGBClassifier:

##Copiar código
#xgb = xgb.XGBClassifier(booster='gbtree')
#Aqui, é criado um objeto do classificador XGBClassifier da biblioteca xgboost. Esse classificador é utilizado para modelos de aprendizado de máquina baseados em árvores de decisão, usando o algoritmo Gradient Boosting.
#O parâmetro booster='gbtree' especifica que o modelo usará árvores de decisão como o "booster" (tipo de modelo base). Outros tipos possíveis de booster são gblinear (para modelos lineares) e dart (Dropouts meet Multiple Additive Regression Trees).
#Realizando a validação cruzada:

##opiar código
##cross_val_score() é uma função da biblioteca sklearn.model_selection que realiza a validação cruzada. Ela avalia o desempenho do modelo em várias divisões do conjunto de dados (em geral, é feito um "k-fold cross-validation").
#xgb é o modelo que será avaliado.
#X_train são os dados de entrada (variáveis independentes), e y_train_encoded são as variáveis alvo (ou seja, as classes ou rótulos).
#O cross_val_score() retorna um array de escores, que são a métrica de avaliação (geralmente a acurácia) para cada divisão do conjunto de dados de validação cruzada.
#O resultado é multiplicado por 100 para transformar os escores em porcentagens, já que o cross_val_score() normalmente retorna valores entre 0 e 1.
#Calculando a pontuação média:

#python
#Copiar código
#average_score = scores.mean()
#O método .mean() calcula a média dos escores obtidos durante a validação cruzada.
#O valor resultante será a média da acurácia ou outra métrica de avaliação que foi usada pelo cross_val_score.
#Exibindo o resultado:

#ython
#Copiar código
#print(f"Pontuação média de validação cruzada: {average_score:.2f}%")
#O resultado final é impresso no formato de uma string formatada.
#{average_score:.2f} significa que o valor da média será exibido com duas casas decimais.
#O print() irá exibir a pontuação média da validação cruzada em porcentagem (pois a multiplicação por 100 foi feita anteriormente).

xgb = xgb.XGBClassifier(booster='gbtree')
scores = 100 * cross_val_score(xgb, X_train, y_train_encoded)
average_score = scores.mean()
print(f"Pontuação média de validação cruzada:{average_score:.2f}%")
```

```python
#Treinamento do modelo:

##Copiar código
#classificador_xgb_gblinear.fit(X_train, y_train_encoded)
#Aqui, o modelo classificador_xgb_gblinear está sendo treinado usando os dados de treinamento X_train e os rótulos codificados y_train_encoded.
#O método .fit() é utilizado para treinar o modelo. Ele recebe dois argumentos:
#X_train: As variáveis independentes (também chamadas de características ou atributos), ou seja, os dados de entrada para o treinamento.
#y_train_encoded: Os rótulos ou classes (variáveis dependentes) que correspondem aos dados em X_train. Neste caso, eles já estão codificados de forma numérica, o que é necessário para o modelo funcionar corretamente.
#O treinamento envolve ajustar os parâmetros do modelo com base nos dados de entrada para aprender a relação entre as variáveis de entrada e os rótulos.
#O classificador_xgb_gblinear provavelmente é uma instância do XGBClassifier com o booster definido como 'gblinear' (em vez de 'gbtree', que é o padrão). Isso significa que o modelo usará gradiente boosting com modelos lineares em vez de árvores de decisão.

#Fazendo previsões com o modelo treinado:

#python
#Copiar código
#predict = classificador_xgb_gblinear.predict(X_test)
#Após o modelo ser treinado, o método .predict() é usado para fazer previsões no conjunto de dados de teste, X_test.
#O modelo utiliza as informações aprendidas durante o treinamento para prever os rótulos ou classes para os dados de teste. X_test contém as mesmas características de entrada que X_train, mas sem os rótulos (que o modelo precisa prever).
#O resultado de .predict() será armazenado na variável predict, que será um array contendo as previsões feitas pelo modelo para cada instância em X_test.

xgb.fit(X_train, y_train_encoded)
predict = xgb.predict(X_test)
```

```python
# Converter os valores preditos para os nomes das classes
predict_labels = le.inverse_transform(predict)

# Criar o DataFrame com as classes reais e preditas
resultado = pd.DataFrame({
    'Real': y_test.values,  # Valores reais já estão com os nomes originais
    'Predito': predict_labels  # Converter para os nomes das classes
})
resultado
```

```python
import pandas as pd
import joblib
from sklearn.preprocessing import LabelEncoder

# 1. Carregar o modelo treinado
modelo_carregado = joblib.load('modelo_premiere02.pkl')

# 2. Carregar os novos dados
dados_novos = pd.read_excel('previsao_premiere_para_teste.xlsx')

# 3. Verificar as colunas presentes no DataFrame
print("Colunas disponíveis no arquivo:", dados_novos.columns)

# 4. Copiar colunas relevantes para referência posterior
times_casa = dados_novos.get('time_casa', None)
times_fora = dados_novos.get('time_fora', None)

# 5. Remover colunas irrelevantes
colunas_irrelevantes = ['Last 5', 'jogador', 'club', 'jogador_expected_goals', 'team', 'data']
dados_novos = dados_novos.drop(columns=[col for col in colunas_irrelevantes if col in dados_novos.columns], errors='ignore')

# 6. Converter colunas categóricas para numéricas usando LabelEncoder
for col in dados_novos.select_dtypes(include=['object']).columns:
    dados_novos[col] = LabelEncoder().fit_transform(dados_novos[col])

# 7. Verificar se todas as colunas estão no formato numérico
print("\nTipos de dados após pré-processamento:")
print(dados_novos.dtypes)

# 8. Garantir que as colunas correspondam ao modelo
colunas_modelo = modelo_carregado.feature_names_in_

# Adicionar colunas faltantes no DataFrame com valor 0
colunas_faltando = [coluna for coluna in colunas_modelo if coluna not in dados_novos.columns]
for coluna in colunas_faltando:
    dados_novos[coluna] = 0

# Remover colunas extras que não são esperadas pelo modelo
dados_novos = dados_novos[colunas_modelo]

# 9. Realizar as previsões
try:
    y_pred = modelo_carregado.predict(dados_novos)
except Exception as e:
    print("\nErro durante a previsão:", str(e))
    raise

# 10. Adicionar as previsões ao DataFrame
dados_novos['Previsão'] = y_pred

# 11. Reintroduzir as colunas originais e a data (se disponíveis)
if times_casa is not None:
    dados_novos['time_casa'] = times_casa
if times_fora is not None:
    dados_novos['time_fora'] = times_fora

# Reintroduzir a coluna 'data', se presente no arquivo original
if 'data' in pd.read_excel('bundesliga_para_teste.xlsx').columns:
    dados_novos['data'] = pd.read_excel('bundesliga_para_teste.xlsx')['data']
else:
    dados_novos['data'] = None

# 12. Mapeamento das previsões
mapeamento_previsoes = {
    0: "vitória_casa",
    1: "empate",
    2: "vitória_visitante"
}
dados_novos['Previsão'] = dados_novos['Previsão'].map(mapeamento_previsoes)

# 13. Exibir as previsões
print("\nPrevisões de resultados para os novos dados:")
for i, (pred, data, time_casa, time_fora) in enumerate(zip(
        dados_novos['Previsão'], 
        dados_novos.get('data', []), 
        dados_novos['time_casa'], 
        dados_novos['time_fora']
    )):
    print(f"Linha {i+1}: Previsão - {pred}, Data - {data}, Time Casa - {time_casa}, Time Fora - {time_fora}")

# 14. Opcional: Salvar as previsões em um arquivo Excel
dados_novos.to_excel('previsoes_bundesliga.xlsx', index=False)
print("\nAs previsões foram salvas em 'previsoes_bundesliga.xlsx'.")
```
