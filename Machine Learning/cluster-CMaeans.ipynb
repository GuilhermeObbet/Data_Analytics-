```python
import numpy as np
from sklearn import datasets
from sklearn.metrics import confusion_matrix
import skfuzzy
```

```python
iris = datasets.load_iris()
```

```python
# Aplicação do algoritmo definido 3 cluster (c = 3) e passando a matriz transposta (iris.data.T) Os outros parametros são obrigatorios e são os defaut indicados na documentação 

r = skfuzzy.cmeans(data = iris.data.T, c = 3, m = 2, error = 0.005, maxiter= 1000, init = None)
```

```python
# Obetendo as porcentagens de um registro pertencer a um cluster, que está na posição 1 da matriz retornada
previsoes_porcentagem = r[1]
```

```python
# Visualização da prob. de um registro pertencer a cada um dis cluster ( o somatorio é 1.0 que indica 100%)

for x in range(150):
  print(previsoes_porcentagem[0][x], previsoes_porcentagem[1][x], previsoes_porcentagem[2][x])
```

```python
# Geração de matriz de contigencia para comparação com as classes originais da base de dados

previsoes = previsoes_porcentagem.argmax(axis = 0)
result = confusion_matrix(iris.target, previsoes)
result
```
